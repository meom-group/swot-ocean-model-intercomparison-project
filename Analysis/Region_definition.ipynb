## Regions defined as in Uchida et al., 2024 where ssh spectra will be computed

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs 
import cartopy.feature as cf
```

```python
xs = [295,318,
      297.5,316,
      300,315
     ]

ys = [37,43,
      30.5,30,
      24,22
     ]
```

```python
x_A = xs[0] -360
x_B = xs[-1] -360
y_A = ys[0]
y_B = ys[-1]

bxA = [x_A-5, x_A+5, x_A+5, x_A-5, x_A-5]
byA = [y_A-5, y_A-5, y_A+5, y_A+5, y_A-5]

bxB = [x_B-5, x_B+5, x_B+5, x_B-5, x_B-5]
byB = [y_B-5, y_B-5, y_B+5, y_B+5, y_B-5]

print(f"Region A limits : {abs(bxA[0])}-{abs(bxA[1])}°W ; {byA[0]} - {byA[2]}°N")
print(f"Region B limits : {abs(bxB[0])}-{abs(bxB[1])}°W ; {byB[0]} - {byB[2]}°N")
```

```python
figure, ax = plt.subplots(1, figsize=(8,6), sharex =True, sharey=True, 
                            subplot_kw={"projection": ccrs.PlateCarree()})

ax.scatter(x_A, y_A, 50, 'k')
ax.scatter(x_B, y_B, 50, 'k')

ax.plot(bxA, byA)
ax.plot(bxB, byB)
ax.coastlines()
gl = ax.gridlines(draw_labels=True)
gl.top_labels=False
gl.right_labels=False
ax.text(-67, 38, "A", fontsize = 15);
ax.text(-47, 23, "B", fontsize = 15);
ax.set_xlim(-85, -30)
ax.set_ylim(15, 50);
```
